```python
def fechar_posicao(position_id, lot, symbol, type_order):
    print("fechar posição")          
    price=mt5.symbol_info_tick(symbol).bid
    deviation=0
    request={
        "action": mt5.TRADE_ACTION_DEAL,
        "symbol": symbol,
        "volume": lot,
        "type": type_order,
        "position": position_id,
        "price": price,
        "deviation": deviation,
        "magic": 234000,
        "comment": "python script close",
        "type_time": mt5.ORDER_TIME_GTC,
        "type_filling": mt5.ORDER_FILLING_RETURN,
    }
    # send a trading request
    result=mt5.order_send(request)
    # check the execution result
    return print("3. close position #{}: {} {} {} lots at {} with deviation={} points".format(position_id,type_order,symbol,lot,price,deviation));
    
```

```python
def abrir_posicao(price_step,price, position_id, lot, symbol, type_order):
        point = mt5.symbol_info(symbol).point        
        #price = mt5.symbol_info_tick(symbol).ask
        price = price
        price_order = 0
        position_id = 0
        #definindo o quanto o valor pode se desviar do valor de compra
        deviation = 0
        if(type_order == mt5.ORDER_TYPE_BUY):
            #definindo o stop loss
            sl = price - 6000 * point            
            round_price_sl = price_step * round(sl/price_step)
            sl = round_price_sl
            #definindo o take profit
            tp = price + 1500 * point            
            round_price_tp = price_step * round(tp/price_step)
            tp = round_price_tp    
        else:
            sl = price + 6000 * point
            print(sl)
            round_price_sl = price_step * round(sl/price_step)
            sl = round_price_sl
            tp = price - 1500 * point
            print(tp)
            round_price_tp = price_step * round(tp/price_step)
            tp = round_price_tp
            
        request = {
            "action": mt5.TRADE_ACTION_DEAL,
            "symbol": symbol,
            "volume": lot,
            "type": type_order,
            "price": price,            
            "sl": sl,            
            "tp": tp,
            "deviation": deviation,
            "magic": 234000,
            "comment": "python script open",
            "type_time": mt5.ORDER_TIME_GTC        
        }
        # send a trading request
        result = mt5.order_send(request)
        print(mt5.last_error())
        # check the execution result
        print("1. order_send(): by {} {} lots at {} with deviation={} points".format(symbol,lot,price,deviation));
        if result.retcode != mt5.TRADE_RETCODE_DONE:
            print("2. order_send failed, retcode={}".format(result.retcode))
           # solicitamos o resultado na forma de dicionário e exibimos elemento por elemento
            result_dict=result._asdict()
            for field in result_dict.keys():
                print("   {}={}".format(field,result_dict[field]))
                #se esta for uma estrutura de uma solicitação de negociação, também a exibiremos elemento a elemento
                if field=="request":
                    traderequest_dict=result_dict[field]._asdict()
                    for tradereq_filed in traderequest_dict:
                        print("       traderequest: {}={}".format(tradereq_filed,traderequest_dict[tradereq_filed]))
                        for trade in tradereq_filed:
                            print("trade" + str(trade))
        print("2. order_send done, ", result)
        print("   opened position with POSITION_TICKET={}".format(result.order))
        t = dict(); 
        t['position_id'] = result.order
        t['price_order'] = result.price        
        print("   sleep 2 seconds before closing position #{}".format(result.order))
        time.sleep(2)
        return t
```

```python
def conferir_hora_operacao():
    hoje = datetime.datetime.now()
    hoje_hora = hoje.hour
    return hoje_hora
```

```python
#usando primeira estrategia de mean reversion
   
request ={}
result = 0
position_id = 0
price_order = 0

#variavel para identificar se a operação aberta está vendida(-1) ou comprada (1)
direction = 0
while True:
    
    timeframe = mt5.TIMEFRAME_M15  # Adjust the timeframe as needed

    # Retrieve historical price data using MetaTrader 5
    data = mt5.copy_rates_from_pos(symbol, timeframe, 0, 500)  # You may adjust the number of bars    
    fechamento = (list(data['close'])[-1])
    # Create a DataFrame from the historical data
    df = pd.DataFrame(data)
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.set_index('time', inplace=True)


    # Calculate the moving average and standard deviation
    df["ma"] = ta.sma(df["close"], length=20)
    df["std"] = ta.stdev(df["close"], length=20)

    # Calculate the z-score
    df["zscore"] = (df["close"] - df["ma"]) / df["std"]
    print("zscore:" + str(df["zscore"].iloc[-1])   )
    
    preco_venda=mt5.symbol_info_tick(symbol).bid
    print("bid: " + str(preco_venda))    
    print("fechamento: " + str(fechamento))  
    
    positions_total = mt5.positions_total()
    print(positions_total)
   
    # Get the value of one tick for the symbol
    tick_value = mt5.symbol_info(symbol).trade_tick_value
    
    # Get the size of the contract for the symbol
    contract_size = mt5.symbol_info(symbol).trade_contract_size
    
    # Get the freeze level for the symbol
    freeze_level = mt5.symbol_info(symbol).trade_freeze_level
    
    # Get the stop level for the symbol
    stop_level = mt5.symbol_info(symbol).trade_stop_level
    
    hoje = conferir_hora_operacao()
    print("hora" + str(hoje))
    #operação das 10h às 17h
    if hoje >= 9 and hoje < 17:
        
        print("horario de operação")        
        # Place the order if the z-score is below -1
        if df["zscore"].iloc[-1] < -1 and positions_total == 0:
            price=df["close"].iloc[-1]
            result = abrir_posicao(tick_value,price, position_id, contract_size, symbol, mt5.ORDER_TYPE_BUY)
            print(result)
            position_id = result["position_id"]
            price_order = result["price_order"]
            print("numero: " +str(position_id))
            print("preco: "+ str(price_order))
            direction = 1
            print(direction)
            
        elif (positions_total > 0):
            # Calculate the profit or loss of the position
            profit = mt5.symbol_info(symbol).trade_calc_profit(position_id, contract_size, price_order, preco_venda)
            print("profit: " + str(profit))
            # Close the position if the profit is positive
            if profit > 0:
                print("fechar posição compra")
                result = fechar_posicao(position_id, contract_size, symbol, mt5.ORDER_TYPE_SELL)
                print(result)
                direction = 0
                print(direction)
            else:
                print("esperar, temos posições abertas")             
        else:
            print("esperar")            
    elif hoje >= 12:
        print("fechar operação")
        zerar_posicao(direction, result)    
    print("")    
    time.sleep(10)
```
